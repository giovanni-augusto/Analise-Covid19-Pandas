```python
# Maior número de mortos em um dia
bd.loc[bd.New_deaths.idxmax(),['Date_reported', 'Country_code', 'Country', 'WHO_region', 'New_cases',
       'Cumulative_cases', 'New_deaths', 'Cumulative_deaths']]
```

```python
# Filtrar a base no último dia disponível
bd_ultimo_dia = bd[bd["Date_reported"] == "2021-08-04"]
bd1 = bd_ultimo_dia.sort_values(by=['Cumulative_cases'], ascending=False)
bd1
```

```python
# Casos acumulados - 04/08/2021
plt.figure(figsize=(15,15))
bd2 = bd1.iloc[0:50]
plt.bar(bd2["Country"],bd2["Cumulative_cases"])
plt.xticks(rotation=90)
plt.title("Casos (total) em 2021-08-04")
plt.show()
```

```python
bd_ultimo_dia = bd[bd["Date_reported"] == "2021-08-04"]
bd3 = bd_ultimo_dia.sort_values(by=['Cumulative_deaths'], ascending=False)
bd3
```

```python
# Mortes totais - 04/08/2021
plt.figure(figsize=(15,15))
bd4 = bd3.iloc[0:50]
plt.bar(bd4["Country"],bd4["Cumulative_deaths"])
plt.xticks(rotation=90)
plt.title("Mortos (total) em 2021-08-04")
plt.show()
```

```python
bd_ultimo_dia = bd[bd["Date_reported"] == "2021-08-04"]
bd5 = bd_ultimo_dia.sort_values(by=['New_deaths'], ascending=False)
bd5
```

```python
# Mortos no dia 04/08/2021
plt.figure(figsize=(15,15))
bd6 = bd5.iloc[0:50]
plt.bar(bd6["Country"],bd6["New_deaths"])
plt.xticks(rotation=90)
plt.title("Número de mortos em 2021-08-04")
plt.show()
```
